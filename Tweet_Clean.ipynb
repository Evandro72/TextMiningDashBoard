```python
import pandas as pd
```

```python
# lendo CSV File
df = pd.read_csv('tweetsAccordiDisaccordi.csv')
df
```

```python
# Tweets baseados nessas Linguas 
# (Note: For some tweets identifying their language was not possible)
Languages = {'ru': 'russian','tr': 'turkish','fr': 'french','ca': 'catalan','hu': 'hungarian','es': 'spanish','no': 'norwegian','en': 'english','it': 'italian','sv': 'swedish','nl': 'dutch','de': 'german','id': 'indonesian','da': 'danish','ro': 'romanian','fi': 'finnish','pt': 'portuguese'}
Languages
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Criando Stopwords Dictionary para cada Language

from nltk.corpus import stopwords

Stop_Words_Dict = {}

for lang in Languages.keys():    
    Stop_Words_Dict[lang] = stopwords.words(Languages[lang].lower())    


Stop_Words_Dict
```

```python
from nltk.corpus import stopwords
from langdetect import detect
```

```python
# Removendo Stopwords

for index, tweet in enumerate(df['Tweet']):
    try:
        # Detecting Languge
        languade_2code = detect(tweet)
        
        df.loc[index, 'Cleaned_Tweet'] = ' '.join([word for word in df.loc[index, 'Tweet'].split() if word not in Stop_Words_Dict[languade_2code]])        
    except:
        # If Language is undetectable
        df.loc[index, 'Cleaned_Tweet'] = df.loc[index, 'Tweet']

df['Cleaned_Tweet']
```

```python
import string, re
```

```python
# Remove Punctuations
df['Cleaned_Tweet'] = df['Cleaned_Tweet'].str.replace('[{}]'.format(string.punctuation), '')
df['Cleaned_Tweet']
```

```python
# Remove Email
df['Cleaned_Tweet'] = [re.sub(r'[A-Za-z0-9]*@[A-Za-z]*\.?[A-Za-z0-9]*', "", str(x)) for x in df['Cleaned_Tweet']]
df['Cleaned_Tweet']
```

```python
# Remove URL
df['Cleaned_Tweet'] = [re.sub(r'^https?:\/\/.*[\r\n]*', '', str(x), flags=re.MULTILINE) for x in df['Cleaned_Tweet']]
df['Cleaned_Tweet']
```

```python
# Remove Mention / Twitter Handler
df['Cleaned_Tweet'] = [re.sub('@[^\s]+', '', str(x)) for x in df['Cleaned_Tweet']]
df['Cleaned_Tweet']
```

```python
# Removendo Special / Unicode Characters
df['Cleaned_Tweet'] = df['Cleaned_Tweet'].str.encode('ascii', 'ignore').str.decode('ascii')
df['Cleaned_Tweet']
```

```python
# Salvado CSV
df.to_csv('Tweets.csv', index=False)
```
